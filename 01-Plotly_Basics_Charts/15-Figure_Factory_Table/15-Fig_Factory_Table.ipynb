# Plotly Figure Factory Table Examples

Plotly's Figure Factory provides a simple way to create formatted tables for data visualization. Unlike `go.Table`, the `create_table()` function automatically handles styling and layout, making it ideal for quick, professional-looking table displays.

```python
# Import the necessary libraries for creating tables and handling data.
import plotly.figure_factory as ff
import pandas as pd
```

## Example 1: Basic Table from List of Lists

This example demonstrates the simplest way to create a table using Figure Factory. By passing a list of lists (where the first row contains headers), you can quickly generate a formatted table without needing to create a DataFrame first. This approach is perfect for small datasets or when you want minimal setup.

```python
# Create a simple table from a list of lists.
# The first row contains column headers, followed by data rows.
data_matrix = [['Country', 'Year', 'Population'],
               ['United States', 2000, 282200000],
               ['Canada', 2000, 27790000],
               ['United States', 2010, 309000000],
               ['Canada', 2010, 34000000]]

# Generate the table using Figure Factory's create_table function.
# This automatically applies default styling and formatting.
fig = ff.create_table(data_matrix)

# Display the table.
fig.show()
```

## Example 2: Table from Pandas DataFrame

When working with data analysis, you'll often have data in a Pandas DataFrame. Figure Factory seamlessly accepts DataFrames and automatically uses the column names as headers. This integration makes it easy to display analysis results or create documentation with your data.

```python
# Create a Pandas DataFrame with sample data.
# DataFrames are ideal for structured data with named columns.
df = pd.DataFrame({
    "Name": ["Luke", "Anakin", "Padme", "Yoda"],
    "Age": [25, 50, 45, 353],
    "Planet": ["Tatooine", "Tatooine", "Naboo", "Dagobah"]
})

# Pass the DataFrame directly to create_table.
# Column names automatically become table headers.
fig = ff.create_table(df)

# Display the table.
fig.show()
```

## Example 3: Customized Table with Color Scale

Figure Factory tables support custom color schemes using the `colorscale` parameter. You can define a gradient by specifying colors at different positions (0 to 1). This is useful for creating branded tables or highlighting data patterns through color. In this example, we use a purple-to-white gradient.

```python
# Create the same data structure as Example 1.
data_matrix = [['Country', 'Year', 'Population'],
               ['United States', 2000, 282200000],
               ['Canada', 2000, 27790000],
               ['United States', 2010, 309000000],
               ['Canada', 2010, 34000000]]

# Apply a custom colorscale to the table.
# The colorscale is defined as a list of [position, color] pairs:
# - Position 0 (start): dark purple (#4d004c)
# - Position 0.5 (middle): light purple (#f2e5ff)
# - Position 1 (end): white (#ffffff)
# This creates a gradient effect across the table rows.
fig = ff.create_table(data_matrix, colorscale=[[0, '#4d004c'],
                                               [0.5, '#f2e5ff'],
                                               [1, '#ffffff']])

# Display the customized table.
fig.show()
```

## Example 4: Table with Custom Font Styling

You can customize the appearance of your table by adjusting font properties. This example shows how to modify font size, family, and color to match your presentation style or branding requirements. Larger fonts improve readability for presentations, while custom colors can help emphasize important information.

```python
# Create a DataFrame with sales data.
df_sales = pd.DataFrame({
    "Product": ["Laptop", "Mouse", "Keyboard"],
    "Price": ["$999", "$29", "$79"],
    "Stock": [45, 230, 156]
})

# Create a table with custom font styling.
fig = ff.create_table(df_sales)

# Customize the font using update_layout.
# This affects all text in the table including headers and data cells.
fig.update_layout(
    font=dict(
        size=14,           # Increase font size for better readability
        family="Arial",    # Use a clean, professional font
        color="darkblue"   # Apply a custom text color
    )
)

# Display the styled table.
fig.show()
```

## Example 5: Table with Custom Height

The `height` parameter allows you to control the vertical size of your table, which is especially useful when you have many rows or want to ensure consistent sizing across multiple visualizations. This helps maintain a clean layout in dashboards or reports.

```python
# Create a DataFrame with employee information.
df_employees = pd.DataFrame({
    "Employee": ["Alice", "Bob", "Charlie", "Diana", "Ethan"],
    "Department": ["Engineering", "Marketing", "Sales", "HR", "Finance"],
    "Years": [5, 3, 7, 2, 4]
})

# Create the table first without the height parameter.
fig = ff.create_table(df_employees)

# Set the height using update_layout after creation.
# The height parameter controls the overall vertical size of the table.
fig.update_layout(height=400)

# Display the table with custom height.
fig.show()
```

## Example 6: Table with Index Column Displayed

By default, Pandas DataFrame index columns are not shown in Figure Factory tables. However, you can include the index by resetting it as a regular column. This is useful when the index contains meaningful information like dates, IDs, or sequential numbers that should be visible in your presentation.

```python
# Create a DataFrame with a meaningful index (customer IDs).
df_orders = pd.DataFrame({
    "Item": ["Tablet", "Phone", "Laptop"],
    "Quantity": [2, 1, 1],
    "Total": [599, 799, 1299]
}, index=["ORD-001", "ORD-002", "ORD-003"])

# Reset the index to convert it into a regular column.
# This makes the index visible in the table as "index" column.
df_with_index = df_orders.reset_index()

# Create the table with the index now displayed as a column.
fig = ff.create_table(df_with_index)

# Display the table showing order IDs.
fig.show()
```
